```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Cargar ambos datasets si no existen en memoria
try:
    df
except NameError:
    df = pd.read_csv("Data/Resultados_BFOA.csv")
try:
    df_mejorado
except NameError:
    df_mejorado = pd.read_csv("Data/Resultados_BFOA_Mejorado.csv")
```

```python
# Comparar estadísticas lado a lado
print("Estadísticas comparativas:")
display(pd.concat([df.describe().T, df_mejorado.describe().T], axis=1, keys=["Original", "Mejorado"]))
```

```python
# Boxplots comparativos
metrics = ["fitness", "tiempo", "blosum", "interaccion"]
for metric in metrics:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=[df[metric], df_mejorado[metric]], palette="Set2")
    plt.xticks([0, 1], ['Original', 'Mejorado'])
    plt.title(f"Comparación de {metric}")
    plt.ylabel(metric)
    plt.grid(True)
    plt.show()
```

```python
# Histogramas superpuestos
for metric in metrics:
    plt.figure(figsize=(8, 5))
    plt.hist(df[metric], bins=15, alpha=0.5, label='Original', edgecolor='black')
    plt.hist(df_mejorado[metric], bins=15, alpha=0.5, label='Mejorado', edgecolor='black')
    plt.title(f"Histograma Comparativo de {metric}")
    plt.xlabel(metric)
    plt.ylabel("Frecuencia")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Scatter comparativo
plt.figure(figsize=(10, 6))
plt.scatter(df["fitness"], df["interaccion"], label="Original", alpha=0.6, edgecolors='black')
plt.scatter(df_mejorado["fitness"], df_mejorado["interaccion"], label="Mejorado", alpha=0.6, edgecolors='red')
plt.xlabel("Fitness")
plt.ylabel("Interacción")
plt.title("Scatter: Fitness vs Interacción (Original vs Mejorado)")
plt.legend()
plt.grid(True)
plt.show()
```
